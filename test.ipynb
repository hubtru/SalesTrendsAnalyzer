# Pivot

```python
import pandas as pd

path = '/home/shady/Projects/AI_lab/Data/merged_cleaned_FE_imputed(v).csv'
df = pd.read_csv(path)

df
```

```python

df_pivot = df.pivot(index="Date", columns="ProductID", values="Quantity")

# sum each row to get total sales for each day and save to a new column total_quantity_day
df_pivot['total_quantity_day'] = df_pivot.sum(axis=1)
df_pivot
```
